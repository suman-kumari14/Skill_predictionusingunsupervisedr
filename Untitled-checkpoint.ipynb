```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
d = load_iris()
X = d.data
wc = []
for i in range(1, 11):
    kmean = KMeans(n_clusters=i, init='k-means++', max_iter=400, n_init=10, random_state=0)
    kmean.fit(X)
    wc.append(kmean.inertia_)
plt.plot(range(1, 11), wc)
plt.title('Elbow Method')
plt.xlabel('Clusters Numbers')
plt.ylabel('WCSS') 
plt.show()
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=400, n_init=10, random_state=0)
kmeans.fit(X)
plt.scatter(X[kmeans.labels_ == 0, 0], X[kmeans.labels_ == 0, 1], s=200, c='blue', label='Cluster 1')
plt.scatter(X[kmeans.labels_ == 1, 0], X[kmeans.labels_ == 1, 1], s=200, c='black', label='Cluster 2')
plt.scatter(X[kmeans.labels_ == 2, 0], X[kmeans.labels_ == 2, 1], s=200, c='orange', label='Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red', label='Centroids')

plt.legend()
plt.show()
```
